```python
# Step 1: Import required libraries
import pandas as pd

# Step 2: Load dataset
df = pd.read_csv('BostonHousing.csv')

# Step 3: Display first few rows and info
print("✅ Dataset Loaded Successfully\n")
print(df.head())
print("\nShape of data:", df.shape)
print("\nColumns:", df.columns.tolist())
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Select feature and target
X = df['rm'].values
y = df['medv'].values

# Drop NaN
X = X[~np.isnan(X)]
y = y[~np.isnan(y)]

# Normalize X
X = (X - np.mean(X)) / np.std(X)

# Initialize parameters
m, c = 0.0, 0.0

# Hyperparameters
L = 0.0001    # very small learning rate
epochs = 5000
n = len(X)

# Gradient Descent
for i in range(epochs):
    y_pred = m * X + c
    D_m = (-2/n) * np.sum(X * (y - y_pred))
    D_c = (-2/n) * np.sum(y - y_pred)
    m -= L * D_m
    c -= L * D_c

print("✅ Training complete!")
print(f"Learned parameters: m = {m:.4f}, c = {c:.4f}")
```

```python
plt.scatter(X, y, color='blue', label='Actual')
plt.plot(X, m*X + c, color='red', label='Predicted Line')
plt.xlabel("Number of Rooms (normalized)")
plt.ylabel("Median House Price (in $1000s)")
plt.title("House Price Prediction using Gradient Descent")
plt.legend()
plt.show()
```
